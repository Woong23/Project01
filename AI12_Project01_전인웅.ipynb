Name : 게임의 이름입니다.
Platform : 게임이 지원되는 플랫폼의 이름입니다.
Year : 게임이 출시된 연도입니다.
Genre : 게임의 장르입니다.
Publisher : 게임을 제작한 회사입니다.
NA_Sales : 북미지역에서의 출고량입니다.
EU_Sales : 유럽지역에서의 출고량입니다.
JP_Sales : 일본지역에서의 출고량입니다.
Other_Sales : 기타지역에서의 출고량입니다.

```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/vgames2.csv')
df = df.fillna(0) #결측치 대체
df.isnull().sum() #결측치 확인

#df1['NA_Sales'].str.contains('K')
#k = df1.loc[df1['NA_Sales'].str.contains('K'),'NA_Sales'] #따로 K 추출
#m = df1.loc[df1['NA_Sales'].str.contains('M'),'NA_Sales'] #따로 M 추출
#k = k.apply(xx) #K 제거 후 1000 나눔
#m = m.apply(yy) #M 제거

def xx(a):  #함수 작성
  if 'K' in a:  #K가 a안에 있으면
    a = a.replace('K','') #K를 제거
    return float(a) / 1000 #a를 부동소수점으로 타입 변경 후 1000으로 나눈다 
  elif 'M' in a:  #M이 a안에 있으면
    a = a.replace('M','') #a에 M제거
    return float(a)       #a 타입 부동소수점으로 변경
  else: 
    return float(a)  #다른 것들은 그냥 그 값 출력

df['NA_Sales'] = df['NA_Sales'].apply(xx) #함수적용
df['EU_Sales'] = df['EU_Sales'].apply(xx) #함수적용
df['JP_Sales'] = df['JP_Sales'].apply(xx) #함수적용
df['Other_Sales'] = df['Other_Sales'].apply(xx) #함수적용
df
```

```python
df1 = df[['Genre','NA_Sales','EU_Sales','JP_Sales','Other_Sales']]
df1.dtypes
df1.groupby('Genre').sum()
```

```python
def yy(b) : #날짜 함수 작성
  if b < 22 :
    return b + 2000
  elif 22 < b < 100:
    return b + 1900
  else :
    return b
df['Year'] = df['Year'].apply(yy) #날짜 함수 적용
df2 = df[['Year','Genre','NA_Sales','EU_Sales','JP_Sales','Other_Sales']] #데이터 프레임 새로 생성
df2 #출력
df2['Year'].describe() #최대 최소 확인

df2.groupby(['Year']).sum()
```
